# wordList(doc)

A function named wordList(doc) in such a way that it takes a txt file as input argument and 
returns a list of words in your document.

```python
# A function wordList(fileName) take fileName as parameter and return slpited list of words
splitedList= []
def wordList(fileName):
    currentFile = open(fileName,'r', encoding = 'Latin-1') 
    content = currentFile.read()    
    #print(content) #it will print original content of file
    global splitedList
    splitedList = content.split()  # The split() breaks the string at the separator and returns a list of strings.
                                #  bydefault separator is whitespace
    currentFile.close()
    #print(splitedList)
    return splitedList
```

# removePuncs(wordList)

A function named removePuncs(wordList) that takes list of words then iterate through this list. During iteration it do some processing on each word. Function should replace punctuation marks as well as \n. and check either this word in stopword on not? if it is in stopword then we didn't append this into resulting List. You must also handle case insensitivity of words.

```python
#a function named removePuncs(wordList) that takes list of words then iterate through this list. 
#During iteration it do some processing on each word. Function should replace punctuation marks as well as \n. 
#and check either this word in stopword on not?
#if it is in stopword then we didn't append this into resulting List. You must also handlevcase insensitivity of words. 
import nltk
nltk.download('stopwords')

def removePuncs(wordList):
    remove_punc=[]
    stop_words = set(stopwords.words('english'))
    for w in wordList:    # iterating in the given list of words
        word=w.translate(str.maketrans({key: None for key in puncs})) # punctuation removal
        word= word.lstrip('\n')
        if word != '':
            remove_punc.append(word) # maintaing a list of punctuation reduced words
    #print(remove_punc)
    final =[]
    for w in remove_punc: #iterating over punctuation removed words list
        new_w = w.lower()  # handling case sensitivity
        if new_w not in stop_words: # eradicating stop words
            final.append(new_w) #maintaing list of final required words
    return final

#testing function's functionality
RemovedPuncs_Corpus = removePuncs(corpus) # calling function removePuncs() : corpus is a list of words obtained as a 
                          #result of previous code block
print(RemovedPuncs_Corpus)
```

# termFrequencyInDoc(wordList)

A function named termFrequencyInDoc(wordList) which should take a list of words as input argument, and output a dictionary of words such that each word that appears in the document is key in the dictionary and it's value is term frequency

```python
def termFrequencyInDoc(wordList):
    tf_dic = {}
    for each in wordList:   #iteration over given list of words
        if each in tf_dic.keys():  #checking if the current word is in dictionary
            tf_dic[each]+=1  #incrementing if current word is already present
        else:
            tf_dic[each]=1  # setting value 1, if current word is not already in dictionary
    return tf_dic    
```
